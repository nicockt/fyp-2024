```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
d = {}
d['Benchmark'] = [0.18, 0.56]
d['Logistic Regression (All Features)'] = [0.14, 0.38]
d['XGBoost (All Features)'] = [0.14, 0.5]
d['SVM (All Features)'] = [0.13, 0.56]
d['LSTM (All Features)'] = [0.0, 0.0]

results_1 = pd.DataFrame.from_dict(d, orient='index')
results_1.columns = ['precision_t', 'recall_t']
print(results_1)

d = {}
d['Logistic Regression (After Feature Engineering)'] = [0.21, 0.56]
d['XGBoost (After Feature Engineering)'] = [0.18, 0.69]
d['SVM (After Feature Engineering)'] = [0.19, 0.54]
d['LSTM (After Feature Engineering)'] = [0.03, 0.5]

results_2 = pd.DataFrame.from_dict(d, orient='index')
results_2.columns = ['precision_t', 'recall_t']
print(results_2)

result = results_1.index.append(results_2.index)
```

```python
# ---------- Plot results recall vs precision ---------- #
colors = ['red', 'blueviolet', 'gold', 'coral', 'green']
colors2 = [ 'blueviolet', 'gold', 'coral', 'green']
markers = ['s', 'o', 'D']

f = lambda m,c: plt.plot([],[],marker=m, color=c, ls="none")[0]
handles = [f('s', colors[i]) for i in range(5)]
handles += [f('o', colors2[i]) for i in range(4)]
labels = list(result)
rcParams.update({'font.size': 16})
rcParams['figure.figsize'] = 20, 16
plt.style.use('fast')

x = list(results_1['recall_t'])
y = list(results_1['precision_t'])
for x_, y_, c_ in zip(x, y, colors):
    plt.scatter(x_, y_, marker='s', c=c_, linewidths=1.5, s=210, alpha=1)

x = list(results_2['recall_t'])
y = list(results_2['precision_t'])
for x_, y_, c_ in zip(x, y, colors2):
    plt.scatter(x_, y_, marker='o', c=c_, linewidths=1.5, s=210, alpha=1)
    
plt.legend(handles, labels, loc=4, framealpha=1)
plt.title('Recall vs Precision all models')
plt.xlim(xmin=0, xmax=1)
plt.ylim(ymin=0, ymax=0.22)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.show()
```
